WYKRES LINIOWY

```python
plt.plot([0,1], [0,3], label='line', color='green')
plt.xlabel('oś x')
plt.ylabel('oś y')
plt.title('Wykres')
_ = plt.legend()
```

```python
x = np.arange(-3, 3,0.1)

x_sin = np.sin(x)
x_cos = np.cos(x)
plt.plot(x, x_sin)
_ = plt.plot(x, x_cos)



```
